```python
import requests
from bs4 import BeautifulSoup
import time

baseurl = 'https://ci.coinafrique.com/categorie/voitures?page='
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

car_data = []
max_items = 100
current_page = 1

print("⏳ Scraping cars...")

while len(car_data) < max_items:
    url = baseurl + str(current_page)
    print(f"\n📄 Page {current_page}: {url}")
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, 'lxml')

    productlist = soup.find_all('div', class_='col s6 m4 l3')

    if not productlist:
        print("⚠️ No more products found.")
        break

    for card in productlist:
        # Car name
        name_tag = card.find('p', class_='ad__card-description')
        name = name_tag.text.strip() if name_tag else "N/A"

        # Price
        price_tag = card.find('p', class_='ad__card-price')
        price = price_tag.text.strip().replace('\n', ' ') if price_tag else "N/A"

        # Seller name
        seller_tag = card.find('p', class_='title')
        seller = seller_tag.text.strip() if seller_tag else "N/A"

        # Location
        location_tag = card.find('p', class_='ad__card-location')
        location = location_tag.text.strip() if location_tag else "N/A"

        # Date posted
        date_tag = card.find('div', class_='card-title ad__card-timesince')
        if date_tag:
            spans = date_tag.find_all('span')
            if len(spans) >= 2:
                date_posted = spans[0].text.strip() + ' ' + spans[1].text.strip()
            else:
                date_posted = "N/A"
        else:
            date_posted = "N/A"

        car_data.append({
            'name': name,
            'price': price,
            'seller': seller,
            'location': location,
            'date_posted': date_posted
        })

        if len(car_data) >= max_items:
            break

    current_page += 1
    time.sleep(1)

# 🎉 Done - Print results
print(f"\n✅ Scraped {len(car_data)} cars:\n")

for i, car in enumerate(car_data, 1):
    print(f"{i}. {car['name']}")
    print(f"   Price      : {car['price']}")
    print(f"   Seller     : {car['seller']}")
    print(f"   Location   : {car['location']}")
    print(f"   Date Posted: {car['date_posted']}\n")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sqlalchemy import create_engine
import pymysql

# 1. SCRAPE DATA
def scrape_coinafrique_cars(pages=3):
    base_url = 'https://www.coinafrique.com/c/vehicules/voitures-occasions?page='
    data = []

    for page in range(1, pages + 1):
        url = base_url + str(page)
        try:
            response = requests.get(url)
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')

            cars = soup.find_all('div', class_='sc-dCFHLb')

            for car in cars:
                title_tag = car.find('span', class_='sc-hknOHE')
                price_tag = car.find('span', class_='sc-llYSUQ')
                location_tag = car.find('span', class_='sc-gLDzan')
                date_tag = car.find('span', class_='sc-cBoprd')

                if title_tag and price_tag:
                    data.append({
                        'title': title_tag.get_text(strip=True),
                        'price': price_tag.get_text(strip=True),
                        'location': location_tag.get_text(strip=True) if location_tag else None,
                        'date_posted': date_tag.get_text(strip=True) if date_tag else None
                    })
        except Exception as e:
            print(f"Failed to scrape page {page}: {e}")

    return pd.DataFrame(data)

# 2. CLEAN DATA
def clean_data(df):
    df['price'] = df['price'].str.replace('[^0-9]', '', regex=True).astype(float)
    df['title'] = df['title'].str.title()
    df.drop_duplicates(inplace=True)
    df.dropna(subset=['title', 'price'], inplace=True)
    return df

# 3. SAVE TO CSV
def save_to_csv(df, filename='car_listings.csv'):
    df.to_csv(filename, index=False)
    print(f"Saved {len(df)} records to {filename}")

# 4. INSERT INTO MYSQL
def insert_into_mysql(df, db_name, table_name, user='root', password='', host='localhost', port=3306):
    try:
        engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{db_name}")
        df.to_sql(name=table_name, con=engine, if_exists='append', index=False)
        print(f"Inserted {len(df)} records into {db_name}.{table_name}")
    except Exception as e:
        print("Failed to insert data into MySQL:", e)

# === MAIN FLOW ===
if __name__ == "__main__":
    df = scrape_coinafrique_cars(pages=3)
    df = clean_data(df)
    save_to_csv(df)

    # Connect and insert into your MySQL database
    insert_into_mysql(
        df,
        db_name='car_data',      # Change to your DB name
        table_name='car_listings',  # Change to your table name
        user='pjct_dwh',
        password='theking',
        host='localhost',        # or '127.0.0.1'
        port=3306
    )
```
